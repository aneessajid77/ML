```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

 

file_path = r"C:\Users\INJETI ANEES SAJID\Downloads\t5_test (1).xlsx"
 

# Reading data from the Excel file into a DataFrame
df = pd.read_excel(file_path)

 

 

# Scatter plot using seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='embed_0', y='embed_1', data=df, alpha=0.6)

 

# Setting title and labels
plt.title('Scatter Plot of embed_7 vs embed_20')
plt.xlabel('embed_7')
plt.ylabel('embed_20')

 

# Display the plot
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

 

# Replace 'independent_feature' and 'dependent_feature' with the actual column names from your DataFrame
# Also, make sure the data types are appropriate

 

# Assuming df is your DataFrame
X = df['embed_0'].values.reshape(-1, 1)
y = df['embed_1'].values

 

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

 

# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

 

# Making predictions
y_pred = model.predict(X_test)

 

# Calculating mean squared error
mse = mean_squared_error(y_test, y_pred)

 

print("Mean Squared Error:", mse)
```
